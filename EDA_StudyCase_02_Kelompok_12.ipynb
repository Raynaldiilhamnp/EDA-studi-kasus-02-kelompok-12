# **Nomor 1**

Apakah ada dosen yang cenderung memberikan nilai cenderung lebih tinggi (atau lebih rendah) dibandingkan dosen lainnya?

```python
nilai_dosen = nilai.groupby(nilai['NIDN']).mean()[['nilai']]
nilai_dosen = nilai_dosen.rename({'nilai': 'mean_nilai'}, axis=1) 
nilai_dosen = pd.merge(dosen, nilai_dosen, left_on='NIDN', right_index=True)
nilai_dosen = nilai_dosen[['NIDN', 'nama', 'mean_nilai']]
sb = nilai.groupby(nilai['NIDN']).std()[['nilai']]
sb = sb.rename({'nilai': 'std_nilai'}, axis=1) 
nilai_dosen = nilai_dosen.set_index(['NIDN']) 
nilai_dosen = pd.concat([nilai_dosen, sb], axis=1)
nilai_dosen
```

```python
a = nilai.merge(dosen, how="outer")
a = a[['NIDN','nama','id_mata_kuliah', 'nilai']]
a = a.groupby([a['NIDN'], a['nama'], a['id_mata_kuliah']]).mean()[['nilai']]
a = a.rename({'nilai': 'mean_nilai'}, axis=1) 
a
b = nilai.merge(dosen, how="outer")
b = b[['NIDN','nama','id_mata_kuliah', 'nilai']]
b = b.groupby([b['NIDN'], b['nama'], b['id_mata_kuliah']]).std()[['nilai']]
b = b.rename({'nilai': 'std_nilai'}, axis=1)
b = pd.concat([a, b], axis=1) 
b
```

# **Nomor 2**

Menurut data yang ada, apakah terdapat perbedaan dalam pemberian nilai antara dosen dengan Pendidikan Terakhir S3 dan S2?

```python
df_penilaian_dosen = pd.merge(nilai, dosen, how="outer")
penilaian_dosen = df_penilaian_dosen[['NIDN', 'nama',  'nilai','Pendidikan_Terakhir', 'id_mata_kuliah']]
penilaian_dosen = penilaian_dosen.groupby([penilaian_dosen['Pendidikan_Terakhir'], penilaian_dosen['id_mata_kuliah']]).mean()[['nilai']]
penilaian_dosen = penilaian_dosen.rename({'nilai': 'mean_nilai'}, axis=1)

std = nilai.groupby(nilai['NIDN']).std()[['nilai']]
std = std.rename({'nilai': 'std_nilai'}, axis=1) 

std_nilai_dosen = df_penilaian_dosen[['NIDN', 'nama',  'nilai','Pendidikan_Terakhir', 'id_mata_kuliah']]
std_nilai_dosen = std_nilai_dosen.groupby([std_nilai_dosen['Pendidikan_Terakhir'], std_nilai_dosen['id_mata_kuliah']]).std()[['nilai']]
std_nilai_dosen = std_nilai_dosen.rename({'nilai': 'std_nilai'}, axis=1)
penilaian_dosen = pd.concat([penilaian_dosen, std_nilai_dosen], axis=1)
penilaian_dosen
```
